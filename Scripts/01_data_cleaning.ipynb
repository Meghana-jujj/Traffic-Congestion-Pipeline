```python
from google.colab import files

uploaded = files.upload()
```

```python
import pandas as pd
filename = list(uploaded.keys())[0]
print(f"Uploaded file: {filename}")

# Step 3: Read CSV file
df = pd.read_csv(filename)

# Step 4: Preview first few rows to check columns
print("Original Data Preview:")
print(df.head())

df_clean = df[['date_time', 'traffic_volume']].copy()

df_clean.rename(columns={
    'date_time': 'timestamp',
    'traffic_volume': 'vehicle_count'
}, inplace=True)

# Add fixed location column
df_clean['location'] = 'I-94 Highway'

# Synthetic avg_speed column: speed decreases as traffic increases
df_clean['avg_speed'] = df_clean['vehicle_count'].apply(lambda x: max(20, 100 - x * 0.1))

# Define congestion level based on vehicle_count
def congestion_level(count):
    if count > 500:
        return 'high'
    elif count > 200:
        return 'medium'
    else:
        return 'low'

df_clean['congestion_level'] = df_clean['vehicle_count'].apply(congestion_level)

output_filename = 'traffic_sample_data_clean.csv'
df_clean.to_csv(output_filename, index=False)
print(f"Cleaned CSV saved as '{output_filename}'")

# Step 7: Preview cleaned data
print("Cleaned Data Preview:")
print(df_clean.head())
```
